```python
#Library import
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import collections 
```

```python
#File Import
G = nx.read_edgelist("protein.edgelist.txt")
```

```python
#Zip deg cnt
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
```

```python
plt.bar(deg, cnt, width = 0.8, color = 'b') #1 Como es el grado de ditribuci√≥n
```

```python
nx.node_connectivity(G) # 2 NODE CONECTIVITY
```

```python
nx.node_connectivity(G,s='4',t='1228') #Para que se desconecte de los nodos 4,1228
```

```python
#Find cliques where n>3
c = zip(*nx.find_cliques(G))
from pandas import DataFrame
data = list(nx.find_cliques(G))
clika = DataFrame.from_records(data)
clika.dropna(thresh=3) #Tabla de cliques > 3
```

```python
#Figure
plt.figure(figsize=(20,20))
#Layout
pos=nx.kamada_kawai_layout(G)
#Drawing in plt using nx
nx.draw_networkx_labels(G, pos = pos, font_size=11, font_family='sans-serif', font_color='black')
nx.draw_networkx_edges(G,pos=pos, alpha=0.005, width=0.1)
nx.draw(G, pos = pos, with_labels = False, node_size =200)
plt.axis('off')
plt.tight_layout();
```
